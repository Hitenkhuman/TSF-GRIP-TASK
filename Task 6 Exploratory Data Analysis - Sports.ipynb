# Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Loading Datasets

```python
match_data=pd.read_csv('matches.csv')
```

# Exploratory Data Analysis

```python
match_data.head(10)
```

```python
#maximum wins by team
plt.figure(figsize=(20,10))
sns.countplot('winner',data=match_data,order=match_data['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Most Wins")
plt.show()
```

```python
#maximum toss winners
plt.figure(figsize=(20,10))
sns.countplot('toss_winner',data=match_data,order=match_data['toss_winner'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
#most player of match awards
plt.figure(figsize=(20,10))
x=match_data['player_of_match'].value_counts()[:10].index
y=match_data['player_of_match'].value_counts()[:10].values
sns.barplot(x = x, y = y)
plt.title("Top 10 player of match award winners ")
plt.xticks(rotation = 90)
plt.show()
```

```python
bat_first=match_data[match_data['win_by_runs']!=0]
bat_first.head(10)
```

```python
#margin by which teams won matches
plt.figure(figsize=(7,7))
plt.hist(bat_first['win_by_runs'])
plt.title("Wins By Bat First")
plt.xlabel('Runs')
plt.show()
```

```python
#winning most match after bat first
plt.figure(figsize=(20,10))
sns.countplot('winner',data=bat_first,order=bat_first['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
bat_second=match_data[match_data['win_by_wickets']!=0]
bat_second.head(10)
```

```python
#winning by margin of wickets
plt.figure(figsize=(7,7))
plt.hist(bat_second['win_by_wickets'],bins=20)
plt.title("Wins By Bat Second")
plt.xlabel('wicket')
plt.show()
```

```python
#winning after bat second
plt.figure(figsize=(20,10))
sns.countplot('winner',data=bat_second,order=bat_second['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
#season wise matches
plt.figure(figsize=(20,10))
x=match_data['season'].value_counts().index
y=match_data['season'].value_counts().values
sns.barplot(x = x, y = y)
plt.title("season wise match ")
plt.xticks(rotation = 90)
plt.show()
```

```python
#winning toss and chose bat first
opt_bat=match_data[match_data['toss_decision']=='bat']
opt_bat.head(10)
```

```python
batfirst_win=opt_bat[opt_bat['toss_winner']==opt_bat['winner']]
```

```python
#winning match by chosing bat first
plt.figure(figsize=(20,10))
sns.countplot('winner',data=batfirst_win,order=batfirst_win['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
#won the toss and select field first
opt_bowl=match_data[match_data['toss_decision']=='field']
opt_bowl.head(10)
```

```python
bowlfirst_win=opt_bowl[opt_bowl['toss_winner']==opt_bowl['winner']]
```

```python
#winning match after chosing bat second
plt.figure(figsize=(20,10))
sns.countplot('winner',data=bowlfirst_win,order=bowlfirst_win['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
#highest score analysis
high_scores=deliveries_data.groupby(['match_id', 'inning','batting_team','bowling_team'])['total_runs'].sum().reset_index() 
high_scores
```

```python
#score is more than 200 runs
score_200=high_scores[high_scores['total_runs']>=200]
score_200
```

```python
plt.figure(figsize=(20,10))
sns.countplot(score_200['batting_team'])
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(score_200['bowling_team'])
plt.show()
```

```python
#maximum runs by bat
max_runs=deliveries_data.groupby(['batsman'])['batsman_runs'].sum()
max_runs.sort_values(ascending=False)[:10].plot.bar()
```
