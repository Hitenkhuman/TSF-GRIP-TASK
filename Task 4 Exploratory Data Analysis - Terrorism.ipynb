# Step:1 Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Step:2 Loading Dataset

```python
data = pd.read_csv('globalTerrorism.csv')
data
```

# Step:3 Data Inspection and Cleaning

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.dropna(axis = 1, how = "all")
```

```python

data.rename(columns={"iyear" : "Year", "imonth" : "Month", "iday" : "Day", "country_txt" : "Country",
                   "region_txt" : "Region", "provstate" : "State", "latitude" : "Latitude", 
                   "longitude" : "Longitude", "summary" : "Summary", "attacktype1_txt" : "AttackType", 
                   "targtype1_txt" : "TargetType", "gname" : "Group", "weaptype1_txt" : "WeaponType", 
                   "nkill" : "Kills", "nwound" : "Wounded",}, inplace=True)
```

```python
data = data[["Year", "Month", "Day", "AttackType", "TargetType", "Country", "Region",
         "Latitude", "Longitude", "Summary", "State", "motive", "WeaponType", 
         "Kills", "Wounded", "Group"]]
data
```

```python
#analyzing the correlation

plt.figure(figsize = (20,10))
sns.heatmap(data.corr(), annot = True)
```

# Step 3:Exploratory Data Analysis

Top 10 Year with the most terrorist attacks

```python
plt.figure(figsize=(10,10))
plt.title('Year')
plt.pie(data['Year'].value_counts()[:10],labels=data['Year'].value_counts()[:10].index,autopct='%1.1f%%')
plt.show()
```


Month with the most terrorist attacks

```python
plt.figure(figsize=(6,6))
plt.title('Month')
plt.pie(data['Month'].value_counts(),labels=data['Month'].value_counts().index,autopct='%1.1f%%')
plt.show()
```


Most terrorist attack type

```python
plt.figure(figsize=(20,10))
sns.countplot(x="AttackType", data=data, order = data["AttackType"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Various attacking methods")
plt.show()
```

Major Terrorist Targets

```python
plt.figure(figsize=(20,10))
sns.countplot(x="TargetType", data=data, order = data["TargetType"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Major Terrorist Targets")
plt.show()
```

Regions Most Affected by Terrorism

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Region", data=data, order = data["Region"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Regions Most Affected by Terrorism")
plt.show()
```

Top 10 Countries most affected by Terrorism

```python
plt.figure(figsize=(20,10))
x=data["Country"].value_counts()[:10].index
y=data["Country"].value_counts()[:10].values
sns.barplot(x = x, y = y)
plt.title("Top 10 Countries most affected by Terrorism")
plt.xticks(rotation = 90)
plt.show()
```

Casualties in Top 15 countries due to Terrorism

```python
plt.figure(figsize=(20,10))

x=data["Country"].value_counts()[:15].index
y=data["Kills"].value_counts()[:15].values
sns.barplot(x = x, y = y)
plt.title("Casualties in Top 15 countries due to Terrorism")
plt.show()
```

Group involved in most terrorist attacks

```python
plt.figure(figsize=(20,10))

x=data["Group"].value_counts()[:5].index
y=data["Group"].value_counts()[:5].values
sns.barplot(x = x, y = y)
plt.title("Group involved in most terrorist attacks")
plt.show()
```
